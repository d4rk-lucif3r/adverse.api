```
%%writefile tutorial/tutorial/spiders/test_india.py
import scrapy
import os
import csv

from bs4 import BeautifulSoup
from unidecode import unidecode
import spacy
from dateutil.parser import parse


nlp_Name = spacy.load("en_core_web_trf") # spacy.load(OUTPUT1)

csv_columns = ['text', 'name']

# defining new variable passing two parameters
writer = csv.DictWriter(open(os.path.abspath(os.path.join(os.getcwd(), 'test_results1.csv')), 'w'), fieldnames=csv_columns)

# writeheader() method to the write to the file object
writer.writeheader()

class IndiaTestSpider(scrapy.Spider):
    name = "test_india"

    def start_requests(self):
        urls = [
                'https://timesofindia.indiatimes.com/india/ed-accuses-amnesty-india-of-laundering-attaches-rs-18cr/articleshow/81022872.cms',
                'https://timesofindia.indiatimes.com/india/ed-attaches-land-in-laundering-case/articleshow/80459638.cms',
                'https://timesofindia.indiatimes.com/india/ed-attaches-over-rs-17cr-assets-of-amnesty-international-india-on-money-laundering-charges/articleshow/80998009.cms',
                'https://timesofindia.indiatimes.com/india/ed-raids-premises-of-news-portal-its-promoters-in-delhi-on-money-laundering-charges/articleshow/80766738.cms',
                'https://timesofindia.indiatimes.com/entertainment/hindi/bollywood/news/armaan-jain-arrives-at-ed-office-in-connection-to-money-laundering-case/articleshow/81039027.cms',
                'https://timesofindia.indiatimes.com/city/mumbai/ed-summons-actor-in-money-laundering-case/articleshow/81006641.cms',
                'https://timesofindia.indiatimes.com/city/mumbai/dawood-mirchi-were-close-met-at-latters-restaurant-witnesses-tell-ed/articleshow/80869270.cms',
                'https://timesofindia.indiatimes.com/city/mumbai/ed-raids-omkar-group-in-money-laundering-case/articleshow/80457039.cms',
                'https://timesofindia.indiatimes.com/city/lucknow/up-ats-arrests-two-chinese-nationals-for-money-laundering/articleshow/80440323.cms',
                'https://timesofindia.indiatimes.com/city/hyderabad/in-2020-ed-hyd-unit-attached-rs-4300-crore-worth-of-assets/articleshow/80696844.cms',
                'https://timesofindia.indiatimes.com/city/varanasi/bsp-leader-killed-over-hawala-biz-rivalry-cops/articleshow/81116013.cms',
                'https://timesofindia.indiatimes.com/city/visakhapatnam/hawala-racket-busted-rs-35l-seized/articleshow/78725762.cms',
                'https://timesofindia.indiatimes.com/city/hyderabad/hawala-gang-busted-four-held-rs-3cr-seized/articleshow/78135579.cms',
                'https://timesofindia.indiatimes.com/world/rest-of-world/cartels-are-scrambling-virus-snarls-global-drug-trade/articleshow/75236173.cms',
                'https://timesofindia.indiatimes.com/india/bihar-resident-arrested-for-terror-links-brought-to-jammu-for-questioning/articleshow/81128429.cms',
                'https://timesofindia.indiatimes.com/india/terror-module-busted-in-jammu-and-kashmir/articleshow/80335694.cms',
                'https://timesofindia.indiatimes.com/india/srinagar-veg-dhaba-scion-hurt-in-terror-attack/articleshow/81079356.cms',
                'https://timesofindia.indiatimes.com/india/terrorists-open-fire-at-police-in-srinagar-one-constable-dead/articleshow/81106772.cms',
                'https://timesofindia.indiatimes.com/world/rest-of-world/cartels-are-scrambling-virus-snarls-global-drug-trade/articleshow/75236173.cms',
                'https://timesofindia.indiatimes.com/india/two-soldiers-injured-in-kashmir-terror-attack/articleshow/80475314.cms',
                'https://timesofindia.indiatimes.com/india/cant-equate-terror-victims-plotters-india-tells-un-body/articleshow/81181297.cms',
                'https://timesofindia.indiatimes.com/india/india-uses-envoys-visit-to-expose-pakistans-terror-role/articleshow/81120162.cms',
                'https://timesofindia.indiatimes.com/india/nia-arrests-jammu-and-kashmir-man-for-helping-hizbul-mujahideen-militants/articleshow/81094730.cms',
                'https://timesofindia.indiatimes.com/india/nia-files-chargesheet-against-ex-jk-cop-who-joined-hizbul-ranks-for-attack-on-crpf-convoy/articleshow/80895938.cms',
                'https://timesofindia.indiatimes.com/india/law-and-order-in-kashmir-improved-a-lot-since-abrogation-of-article-370-provisions-crpf-officer/articleshow/80912893.cms',
                'https://timesofindia.indiatimes.com/india/jem-planned-terror-attacks-in-delhi-terrorists-sourcing-weapons-from-bihar-jk-dgp/articleshow/80911547.cms',
                'https://timesofindia.indiatimes.com/india/major-attack-averted-in-jk-on-pulwama-anniversary-6-5kg-ied-recovered/articleshow/80909684.cms',
                'https://timesofindia.indiatimes.com/india/pulwama-terror-attack-rahul-slams-pm-for-his-ignorance-on-intelligence-inputs/articleshow/80983439.cms',
                'https://timesofindia.indiatimes.com/india/jammu-and-kashmir-issue-can-only-be-settle-through-talks-between-india-pak-mehbooba/articleshow/81137261.cms',
                'https://timesofindia.indiatimes.com/india/terrorist-groups-their-handlers-trying-to-shift-focus-from-kashmir-valley-to-jammu-jitendra-singh/articleshow/81129365.cms',
                'https://timesofindia.indiatimes.com/india/terror-hideout-busted-in-jks-poonch-huge-cache-of-arms-seized/articleshow/80421134.cms',
                'https://timesofindia.indiatimes.com/business/india-business/indiabulls-housing-eyes-200mn-via-fccb-issue/articleshow/81160444.cms',
                'https://timesofindia.indiatimes.com/india/routine-for-pakistan-to-come-up-with-farcical-actions-india-questions-timing-of-action-against-mumbai-attack-mastermind-lakhvi/articleshow/80174019.cms',
                'https://timesofindia.indiatimes.com/world/pakistan/pak-anti-terrorism-court-jails-hafiz-saeeds-3-aides-in-terror-financing-case/articleshow/80422423.cms',
                
        ]
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
      # text data of entire webpage
      soup = BeautifulSoup(response.text, 'html.parser') # parse
      # extract body from it
      tag = soup.body # scrape.parse_soup
      
      # save data in profile
      profile = {'text': '', 'name': '', 'org': '', 'loc': '', 'facebook_link': '', 'twitter_link': '', 'other_weblinks': '', 'image_link': '', 'keyword': '', 'sources': '', 'hdfcpresent': 'No', 'date': ''}

      fp_name = [
                 'Shiv sena'
                 'Gaana',
                 'Mensxp',
                 'Maharashtra Corona cases',
                 'Bigg Boss',
                 'Gujarati',
                 'Wonder Woman 1984',
                 'Andhaghaaram',
                 'Suraj Pe Mangal Bhari',
                 'Chhalaang',
                 'Ludo',
                 'Laxmii',
                 'Soorarai Pottru',
                 'Miss India',
                 'Varmaa',
                 'Oriya',
                 'Tulu',
                 'Chakra Ka Rakshak',
                 'Meera Mathur',
                 'Sakshi',
                 'Chhote Bade Bhaijaan',
                 'Maassab',
                 'Satya Sai Baba',
                 'moreli',
                 'Corona cases in Mumbai',
                 'delhi',
                 'ajmer',
                 'amaravati',
                 'amritsar',
                 'dehradun',
                 'hubballi',
                 'itanagar',
                 'kochi',
                 'kohima',
                 'ludhiana',
                 'madurai',
                 'mangaluru',
                 'ranchi',
                 'thane',
                 'shimla',
                 'varanasi',
                 'vijayawada',
                 'FCRA',
                 'Femina',
                 'ETimes',
                 'Grazia',
                 'RTI Bill',
                 'The Economic Times',
                 'Hot on the Web',
                 'Corona cases in India',
                 'EMail',
                 'Jharkhand',
                 'Gujarat',
                 'Assam',
                 'Haryana',
                 'Mizoram',
                 'Nagaland',
                 'Odisha',
                 'Sikkim',
                 'Tripura',
                 'Dadra Nagar Haveli',
                 'Daman Diu',
                 'Lakshadweep',
                 ]

      fp_org = [

                 ]

      fp_loc = [


                 ]
      
      fp_date = [


                 ]

      keywords = ['money laundering', 
                  'money launder', 
                  'lauder the money', 
                  'money-mule', 
                  'money mule', 
                  'hawala', 'drug-trafficking', 'drug trafficking', 'terror', 'terror financing']
      # iterate through each string
      for string in tag.stripped_strings:

        for keyword in keywords:
          if keyword in str(string).lower():
            if keyword not in profile['keyword']:
              profile['keyword'] += keyword + ', '
        
        if 'HDFC' in str(string).lower():
          profile['hdfcpresent'] = 'Yes'


        if len(string) > 540:
          continue
        
        # name extration through spacy model
        doc = nlp_Name(str(string))
        
        # iterate through each entity present
        for count,ent in enumerate(doc.ents):
          # print(ent.label_, ':', ent.text)
          
          # find persons in text
          if ent.label_ == 'PERSON':
            
            # remove name if present in false positives
            if str(string) not in fp_name and (ent.text not in profile['name']):
              # print(str(string))
              profile['text'] = str(string)
              profile['name'] += ent.text + ', '
            else:
              # print('Persons entity:', ent.text, ':', ent.label_)
              pass
          
          # find persons in text
          elif ent.label_ == 'ORG':
            
            # remove name if present in false positives
            if str(string) not in fp_org and (ent.text not in profile['org']):
              profile['org'] += ent.text + ', '
            else:
              # print('Persons entity:', ent.text, ':', ent.label_)
              pass

          # find persons in text
          elif ent.label_ == 'GPE':
            
            # remove name if present in false positives
            if str(string) not in fp_loc and (ent.text not in profile['loc']):
              profile['loc'] += ent.text + ', '

          # find persons in text
          elif ent.label_ == 'DATE':
            
            # remove name if present in false positives
            if str(string) not in fp_date and (ent.text not in profile['date']) and (not (ent.text).isalpha()) and (not (ent.text).isdigit())and (not profile['date']):
              try:
                profile['date'] += str(parse(ent.text)) + ', '
              except:
                pass
          
      
      profile['sources'] = response.url

      yield profile
```
